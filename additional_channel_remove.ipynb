```python
def removeBadChanel(path, folder):

  for i in range(len(path)):
    image = plt.imread(path[i], 0)
    try:
      if len(image.shape) == 3:
        if image.shape[2] == 3:
          image = img_as_ubyte(image)
          imageio.imwrite('/content/drive/My Drive/JOGA_DATASET/test/' + folder + '/'+ str(i) +'.jpg' , image)
    except Exception as e:
      print(e)
```

```python
removeBadChanel(train_tree, 'tree')
removeBadChanel(train_plank, 'plank')
removeBadChanel(train_downdog, 'downdog')
removeBadChanel(train_warrior2, 'warrior2')
removeBadChanel(train_goddess, 'goddess')
```

```python
removeBadChanel(test_tree, 'tree')
removeBadChanel(test_plank, 'plank')
removeBadChanel(test_downdog, 'downdog')
removeBadChanel(test_warrior2, 'warrior2')
removeBadChanel(test_goddess, 'goddess')
```
